```python
import pandas as pd
from openpyxl import load_workbook
from openpyxl.styles import PatternFill

# =========================
# ① Excel読み込み
# =========================
file_2022 = r"2022_年間売上表.xlsx"
file_2023 = r"2023_年間売上表.xlsx"

df_2022 = pd.read_excel(file_2022, sheet_name="Sheet1")
df_2023 = pd.read_excel(file_2023, sheet_name="Sheet1")

# =========================
# ② データ結合
# =========================
df_all = pd.concat([df_2022, df_2023], ignore_index=True)

# =========================
# ③ 商品×売上年で集計
# =========================
df_summary = (
    df_all
    .groupby(["商品", "売上年"], as_index=False)["金額（千円）"]
    .sum()
)

# 並び替え（画像と合わせる）
df_summary = df_summary.sort_values(["商品", "売上年"])

# =========================
# ④ Excel書き込み
# =========================
output_file = r"売上集計表.xlsx"
df_summary.to_excel(output_file, index=False)

# =========================
# ⑤ ヘッダーを薄いグレーに変更
# =========================
wb = load_workbook(output_file)
ws = wb.active

# 薄いグレー（#F2F2F2）
header_fill = PatternFill(start_color="F2F2F2",
                          end_color="F2F2F2",
                          fill_type="solid")

# 1行目（ヘッダー）に適用
for cell in ws[1]:
    cell.fill = header_fill

# 保存
wb.save(output_file)

print("売上集計表.xlsx を作成しました。")
```
