```python
from openpyxl import Workbook
from openpyxl.styles import Font, Alignment
from datetime import datetime

# ワークブック作成
wb = Workbook()
ws = wb.active
ws.title = "請求書"

# ===== フォント設定 =====
title_font = Font(size=16, bold=True)
bold_font = Font(bold=True)

# ===== タイトル =====
ws["B2"] = "請求書"
ws["B2"].font = title_font
ws["B2"].alignment = Alignment(horizontal="center")

# ===== 固定データ =====
ws["B4"] = "株式会社ABC"
ws["B5"] = "〒101-0022 東京都千代田区神田練塀町300"
ws["B6"] = "TEL:03-1234-5678 FAX:03-1234-5678"
ws["B7"] = "担当者名:鈴木一郎 様"

# ===== 請求書情報 =====
ws["F4"] = "No."
ws["G4"] = "0001"

ws["F5"] = "日付"
ws["G5"] = datetime(2026, 3, 21)
ws["G5"].number_format = "yyyy/mm/dd"

# ===== ヘッダー =====
ws["B10"] = "商品名"
ws["C10"] = "数量"
ws["D10"] = "単価"
ws["E10"] = "金額"

for cell in ["B10", "C10", "D10", "E10"]:
    ws[cell].font = bold_font
    ws[cell].alignment = Alignment(horizontal="center")

# ===== 明細 =====
data = [
    ["商品A", 2, 10000, 20000],
    ["商品B", 1, 15000, 15000],
]

start_row = 11

for i, row in enumerate(data):
    ws.cell(row=start_row + i, column=2, value=row[0])
    ws.cell(row=start_row + i, column=3, value=row[1])
    ws.cell(row=start_row + i, column=4, value=row[2])
    ws.cell(row=start_row + i, column=5, value=row[3])

# ===== 合計計算 =====
subtotal = sum(row[3] for row in data)
tax = int(subtotal * 0.1)
total = subtotal + tax

ws["D15"] = "合計"
ws["E15"] = subtotal

ws["D16"] = "消費税"
ws["E16"] = tax

ws["D17"] = "税込合計"
ws["E17"] = total

# ===== 列幅調整 =====
ws.column_dimensions["B"].width = 20
ws.column_dimensions["C"].width = 10
ws.column_dimensions["D"].width = 15
ws.column_dimensions["E"].width = 15
ws.column_dimensions["F"].width = 10
ws.column_dimensions["G"].width = 15

# ===== 保存 =====
file_name = "請求書_20260321.xlsx"
wb.save(file_name)

print(f"{file_name} を作成しました。")
```
